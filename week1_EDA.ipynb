## 4. Feature 생성

## 1. 시간

```python
# datetime 형태로 바꿔주기 
datelist = ['Appraisal_date', "First_auction_date", "Final_auction_date", "Preserve_regist_date", "Close_date"]

for date in datelist: 
    data[date] = pd.to_datetime(data[date], errors = 'coerce')
```

```python
data[['Appraisal_date', "First_auction_date", "Final_auction_date", "Preserve_regist_date", "Close_date"]].head()
```

### 1) 경매날짜의 년도&월
* Close_date: 종국일자 

```python
data["Close_date_year"] = data['Close_date'].dt.year
data["Close_date_month"] = data['Close_date'].dt.month
```

```python
figure, axes = plt.subplots(nrows=2)
figure.set_size_inches(10,8)

sns.barplot(data['Close_date_year'], data['Hammer_price'], ax=axes[0])
sns.barplot(data['Close_date_month'], data['Hammer_price'], ax=axes[1])
```

* 해가 지날수록 큰 차이는 없는 것 같지만, 월마다는 차이가 있어 보임
* 3월, 4월 낙찰가가 좀 높아 보임...

### 2) 경매날짜의 분기 (1-4분기)
* Close_date: 종국일자

```python
def season(x):
    if x in [1,2,3]:
        return "1분기"
    elif x in [4,5,6]:
        return "2분기"
    elif x in [7,8,9]:
        return "3분기"
    elif x in [10,11,12]:
        return "4분기" 
```

```python
data["Close_date_season"] = data['Close_date_month'].apply(season)

data["Close_date_season"] = pd.Categorical(data["Close_date_season"], 
                                           categories = ["1분기", "2분기", "3분기", "4분기"], ordered = True)
```

```python
figure.set_size_inches(10,3)

sns.barplot(data["Close_date_season"] , data['Hammer_price'])
```

* 분기마다 미세한 차이를 보임 
* 4분기에는 낙찰가가 타 분기보다 좀 낮게 형성되어 있는 편 

### 3) 경매 기간 
* First_auction_date 최초경매일
* Final_auction_date 최종경매일

```python
data["Auction_duration"] = data["Final_auction_date"]-data["First_auction_date"]
```

```python
data3 = data.loc[data["Auction_duration"].dt.days<500]
```

```python
figure, axes = plt.subplots(ncols=2)
figure.set_size_inches(12,5)

sns.scatterplot((pd.Series(data["Auction_duration"]).dt.days), data['Hammer_price'], color="blue", ax=axes[0])
sns.scatterplot((pd.Series(data3["Auction_duration"]).dt.days), data3['Hammer_price'], color="red", ax=axes[1])
```

* (최종 - 최초): 매각을 실시하고, 매각을 결정하기까지의 기간
* 두번째 그림은 첫번째 데이터에서 (auction_duration<500)인 부분만 좀 더 확대해서 본 것

### 10) 용적률
* 용적률 = (건물면적 x 층수) / 대지면적 
* Total_floor: 총층수
* Total_land_auction_area: 총토지경매면적(㎡) - 건물면적
* Total_building_auction_area: 총건물경매면적(㎡) - 대지면적

* 용적률이 클수록 대지면적에 비해서 건축물 연면적의 비율이 높다는 것
* 용적률이 크다는 것은 그만큼 건물을 높게 지었다 즉, 건물의 층수가 많다는 것을 의미함
* 따라서 용적률이 크면 그만큼 건물을 여러 층으로 높게 지을 수 있음 
<br> 


* 용적률 또한 토지의 가격에 직접적인 영향을 미침: 용적률이 높은 토지가 용적률이 낮은 토지에 비해 가격이 높게 형성됨
